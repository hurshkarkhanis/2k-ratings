## Data Cleaning

```python
# import pandas
import pandas as pd
pd.set_option('display.max_columns', None)

# import beautiful soup to web scrape 2k ratings from hoopehype.com
from bs4 import BeautifulSoup
import requests

# import matplotlib for visualizations
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
# players.csv is a dataset containing their name as well as their total statistics from the 2019-2020 season
# total statistics include total points, total rebounds, total assists, field gola percentage, etc etc
players = pd.read_csv('players.csv')
players = players.dropna(axis=0)
players.columns = [x.lower() for x in ['rk', 'player', 'pos', 'age', 'tm', 'g', 'gs', 'mp', 'fg', 'fga',
       'fg%', '3p', '3pa', '3p%', '2p', '2pa', '2p%', 'efg%', 'ft', 'fta',
       'ft%', 'orb', 'drb', 'trb', 'ast', 'stl', 'blk', 'tov', 'pf', 'pts']]
```

```python
player_name = []
site_name = []
```

```python
# function to take a players name and convert in a way that will fit into the hoopshype.com link
# ex: Marvin Bagley III gets converted to marvin-bagley for link hoopshype.com/player/marvin-bagley/2k/
import unidecode
def name_for_site(before, location):
    
    #removing any 'III' or 'II' at end of name
    # example "Marvin Bagley III"
    if before.endswith(' III'):
        before = before.split(' III')[0]

    if before.endswith(' II'):
        before = before.split(' II')[0]
        
    if before.endswith(' Jr.'):
        before = before.split(' Jr.')[0]  
        
        
    out = ''
    
    for x in before:
        if x == '.':
            out += ''    
        elif x != ' ':
            out += x.lower()
        else:
            out +='-'
    out = unidecode.unidecode(out)
    
    # namy names are 'one off' cases that I couldnt standardzie, so I had to hard code them.
    # not proud of this, but couldn't think of another way
    if out == 'jj-barea':
        out = 'jose-juan-barea'
    
    if out == 'jaren-jackson':
        out = 'jaren-jackson-jr'
    if out == 'timothe-luwawu-cabarrot':
        out = 'timothe-luwawu'
    if out == 'patty-mills':
        out = 'patrick-mills'
    if out == 'larry-nance':
        out = 'larry-nance-jr'
    if out == 'dennis-schroder':
        out = 'dennis-schroeder'
    if out == 'ish-smith':
        out = 'ishmael-smith'
    if out == 'edmond-sumner':
        out = 'edmund-sumner'
    if out == 'gary-trent':
        out = 'gary-trent-jr'
    if out == 'lou-williams':
        out = 'louis-williams'
    if out == 'tim-hardaway':
        out = 'tim-hardaway-jr'
    if out == 'maurice-harkless':
        out = 'moe-harkless'
    if out == 'luc-mbah-a-moute':
        out = 'luc-richard-mbah-a-moute'
    if out == 'glenn-robinson':
        out = 'glenn-robinson-iii'
        

        
    location.append(out)
```

```python
for x in players.player:
    player_name.append(x.split("\\")[0])
    
for x in player_name:
    name_for_site(x,site_name)
```

```python
players.player = site_name
```

```python
players = players.drop(['rk'], axis=1)
```

## Web Scraping

```python
def get_rating(ID):

    # namy names are 'one off' cases that I couldnt standardzie, so I had to hard code their ratings.
    # not proud of this, but couldn't think of another way
    if ID == 'jaren-jackson-jr':
        return 83
    if ID == 'chris-silva':
        return 73
    if ID == 'charlie-brown':
        return 70
    if ID == 'gary-payton':
        return 72
    if ID == 'juan-toscano-anderson' or ID == 'nigel-williams-goss':
        return 69
    
    # this is the link used to grab 2k ratings
    url = 'https://hoopshype.com/player/{}/2k/'.format(ID)
    r  = requests.get(url)
    data = r.text
    soup = BeautifulSoup(data, "html.parser")
    mydivs = soup.findAll('td')[5]
    
    return int(mydivs.contents[0])
```

```python
ratings = []
for x in players.player:
    ratings.append(get_rating(x))
```

## Readying Data for Linear Regression

```python
players['2k_rating'] = ratings
```

```python
# dummifiying position and team data
position_dummies = pd.get_dummies(players.pos)
team_dummies = pd.get_dummies(players.tm)
```

```python
frames = [players.drop(['tm', 'pos'], axis=1), position_dummies, team_dummies]
```

```python
ready_for_model = pd.concat(frames, axis=1)
```

```python
# dataset ready for linear regression
ready_for_model = ready_for_model.drop(['player'], axis = 1)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, KFold
import sklearn.metrics as sk
```

```python
X = ready_for_model.drop(['2k_rating'], axis =1)
y = ready_for_model['2k_rating']

# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# function to run linear regression with default paremeters and graph true vs predicted values
def regression_model(X, y, my_model):
    model = my_model
    model.fit(X_train, y_train)
    predicted = model.predict(X_test)
    rmse = sk.mean_squared_error(y_test,predicted)

    
    real = y_test.tolist()
    pred = predicted.tolist()
    x = list(range(0, len(real)))
    
    fig, ax = plt.subplots(figsize=(20, 5))
    ax.plot(x, real, label='actual', color='black', marker="X")
    ax.plot(x, pred, label='predicted', color='red', marker="X")
    ax.set_title("NBA 2K21 Player Ratings - actual vs predicted. RMSE: {}".format(rmse))
    ax.set_xlabel("Prediction #")
    ax.set_ylabel("NBA 2K21 Rating")
    
    
    ax.legend()    
```

## Results

```python
# running above function
regression_model(X, y, LinearRegression())
```

```python
#feature importance on which statistical categories most influenced a 2k rating
model = LinearRegression()
model.fit(X_train, y_train)
predicted = model.predict(X_test)

from sklearn.inspection import permutation_importance
r = permutation_importance(model, X, y, n_repeats=30,random_state=0)

feature_importance = {}

for i in r.importances_mean.argsort()[::-1]:
     if r.importances_mean[i] - 2 * r.importances_std[i] > 0:
        feature_importance[f"{ready_for_model.columns[i]:<8}"] = f"{r.importances_mean[i]:.3f}"
```
